```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x,y,N):
    x_sum = 0
    y_sum = 0
    x_sq = 0
    xy = 0
    for i in range(N):
        x_sum += x[i]   #a
        y_sum += y[i]   #b
        x_sq += pow(x[i],2) #c
        xy += x[i]*y[i]     #d
    return [x_sum, y_sum, x_sq, xy]

def coef(a,b,c,d,N):
    A = np.array([[N,a],[a,c]])
    inv_A = np.linalg.inv(A)
    B = np.array([b,d])
    X = np.linalg.inv(A).dot(B)
    return X

def SSE(x,y,w0,w,N):    #The sum of squares of residuals
    e = 0
    for i in range(N):
        e += pow(y[i]-w*x[i]-w0,2)
    return e

def SST(x,y,N):     #The total sum of squares
    y_ave = np.mean(y)
    e = 0
    for i in range(N):
        e += pow(y[i]-y_ave,2)
    return e
```

```python
N = 10
x = [1, 1, 2, 2, 2, 3, 3, 4, 5, 6]
y = [1, 2, 2, 3, 4, 4, 5, 5, 6, 7]
s = function(x,y,N)
w = coef(s[0],s[1],s[2],s[3],N)
e1 = SSE(x,y,w[0],w[1],N)
e2 = SST(x,y,N)
print("y = ", np.around(w[1],4),"x + ", np.around(w[0],4))
print("R squared = ", np.around(1-e1/e2,4))
x1 = np.linspace(1,6,N)
result = w[1]*x1 + w[0]
```

```python
plt.plot(x,y,'o',label = "Data")
plt.plot(x1,result,label = "linear regression",color = 'red')
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()
```
