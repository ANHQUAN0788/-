```python
import numpy as np
import matplotlib.pyplot as plt
```

## 関数定義

```python
# 正規化処理
def z_sore(x, ave, var):

  y = (x - ave)/np.sqrt(var)

  return y

# 信号の定義
def signal(t, fs, f):
  y = np.sin(2*np.pi*f*t) + 1/2*t - 5
  return y
```

##実験データ作成

サンプリング周波数4Hzである。
使用する信号は周波数2Hzのsin波と一次関数の合成信号

３秒間の信号を処理x軸時間、y軸振幅とする

振幅を平均0、分散1とする処理にする  

```python
import numpy as np

# 変数定義
fs =  40    # サンプリング周波数
f =   2    # 信号の周波数
t =   3    # 時間


# 0 ～t秒までサンプリング周波数の刻みごとのndarrayを作成
t = np.arange(0,t,1/fs)

#処理前の信号の作成
before_sg = signal(t, fs, f)
```

## 正規化処理と比較

```python
#　必要な値の取得
mean = np.mean(before_sg,axis = None)
var = np.var(before_sg,axis = None)
after_sg = z_sore(before_sg, mean, var)

print('var_after =',np.var(after_sg))
plt.plot(t, before_sg,label="before")
plt.plot(t, after_sg,label="after")

plt.legend()
plt.xlabel("Time[s]", fontsize=12)
plt.ylabel("Amplitude", fontsize=12)
plt.grid()
plt.show()
```
