```python
from numpy.core.function_base import linspace
#　一つのトライアルデータの作成
# 入力は 時間(ndarray)、サンプリング周波数、周波数（信号ごとにランダムで周波数が変わるようにする）
def signal(t, fs, rand_f):
  # 信号ごとに周波数が変わるように
  rand_f =  rand_f + np.random.randn(fs*t1)*0.4
  # 信号の作成（正弦波でもなんでも）
  y = np.sin(2*np.pi*rand_f*t)

  # 振幅にノイズ成分を追加する
  y = y + np.random.randn(t1*fs)*0.5

  return y


# 複数のトライアルデータの作成
def create_data(trial, t, fs, f):

  data = np.array([])

  # 　ndarrayに各トライアルデータを追加
  for i in range(1):
    data = np.append(data,trial)

  return data

# 変数定義
fs = 40    # サンプリング周波数
f =  1    # 信号の周波数
t1 = 3      
t = np.arange( 0 , t1, 1/fs) # 時間
# 1～９番目のトライアルのみ確認
fig = plt.figure(figsize=(20, 16))
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 12

for i in range(9):
  trial = signal(t, fs, f)
  data = create_data(trial, t, fs, f)
  ax = fig.add_subplot(3,3,i+1)
  ax.plot(t,data)

  plt.xlabel("Time", fontsize=12)
  plt.ylabel("Signal", fontsize=12)
  plt.title('traial'+str(i+1))
  plt.grid()
  leg = plt.legend(loc=1, fontsize=15)
  leg.get_frame().set_alpha(1)

#　図同士がかぶらないようにする
```
