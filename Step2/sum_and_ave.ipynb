# 加算平均

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 関数の定義

```python
def sum_average(x,t,f,fs):
  N  = fs*3
  y = x[0]
  print(y)
  for i in range(N):
    y += np.sin(2*np.pi*f*t)+ np.random.randn(N)
  x_mine = y/(N)
  return x_mine

def signal(t, fs, f):
  N = fs*t
  # 信号の作成（正弦波でもなんでも）
  t = np.arange(0,t,1/fs)
  y = np.sin(2*np.pi*f*t)
  y = y + np.random.randn(N)*0.5
  return y 
def create_data(trial, t, fs, f):

  data = np.array([])

  # 　ndarrayに各トライアルデータを追加
  for i in range(1):
    data = np.append(data,trial)

  return data
```

##  加算平均

```python
fs = 40 # サンプリング周波数
f = 1/2
trial = signal(3,fs,f) 

# 0 ～t秒までサンプリング周波数の刻みごとのndarrayを作成
t = np.arange(0,3,1/fs)
before_data = create_data(trial, t, fs, f)
after_data = sum_average(before_data,t,f,fs)
```

## 描画

```python
plt.plot(t,before_data,label="before")
plt.plot(t,after_data,label="after")
plt.legend()
plt.xlabel("Time[s]")
plt.ylabel("Amp")
plt.grid()
plt.show()
```
